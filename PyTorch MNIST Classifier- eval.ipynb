```python
import torch
import torchvision
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import pickle
```

```python
def evaluate_hw1():
  # Hyper Parameters
  input_size = 784
  num_epochs = 10
  batch_size = 100

  transform = transforms.Compose(
          [transforms.ToTensor(),
          transforms.Normalize((0.1307,), (0.3015,))])

  # Dataset Loader (Input Pipline)
  test_dataset = dsets.MNIST(root='./data',
                           train=False,
                           transform=transform)

  test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                            batch_size=batch_size,
                                            shuffle=False)

  #load my trained network
  nn = pickle.load(open("train_hw1_206238891.pkl", 'rb'))

  #calculate avg accuracy of the test set
  test_correct = 0
  test_total = 0
  test_count =[]
  test_acc =[]

  for i, (test_images, test_labels) in enumerate(test_loader):

    test_images= test_images.view(-1, 784)
    one_hot_test= torch.zeros(test_labels.size(0), 10)
    one_hot_test[torch.arange(test_labels.size(0)), test_labels]=1

    #prediction = forward step
    y_pred_test = nn.forward(test_images)
    _, predicted_test = torch.max(y_pred_test.data, 1)

    test_correct += ((predicted_test == test_labels).sum())
    test_total += test_labels.size(0)
    test_acc.append(test_correct/ test_total)
    test_count.append(test_total)

  plt.plot(test_count,test_acc , color = 'green')
  plt.title("Accuracy Value")
  plt.xlabel("Number Of Samples")
  plt.ylabel("Accuracy")
  plt.show()

  print ('Accuracy of the Test Set is: %f %%' % (100 * test_correct / test_total))
```

```python
evaluate_hw1()
```
